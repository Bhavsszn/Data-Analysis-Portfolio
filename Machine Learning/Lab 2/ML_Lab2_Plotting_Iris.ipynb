## Importing Libraries and Packages

## Imports

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Example 1: Visualizing Time vs Distance Relationships

```python
# Line Plot: Time vs Distance
time = [0, 1, 2, 3, 4, 5]
distance = [0, 30, 60, 90, 120, 150]
plt.plot(time, distance)
plt.xlabel('Time (in hours)')
plt.ylabel('Distance (in miles)')
plt.title('Line Plot: Time vs Distance')
plt.show()

# Bar Plot: Time vs Distance
plt.bar(time, distance)
plt.title('Bar Plot: Time vs Distance')
plt.show()

# Scatter Plot: Time vs Distance
plt.scatter(time, distance)
plt.title('Scatter Plot: Time vs Distance')
plt.show()
```


## Custom Visualizations for Mathematical Functions

```python
# Line and Bar Plot for Quadratic Growth
x = np.arange(0, 10, 1)
y = x ** 2
plt.plot(x, y, linestyle='--', color='purple')
plt.title('Line Plot: Quadratic Growth')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

plt.bar(x, y, color='orange')
plt.title('Bar Plot: Quadratic Growth')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Analysis of Iris Dataset Features

```python
# Line and Bar Plot for Petal Length and Sepal Width
iris_data = pd.read_csv("iris.csv")

plt.plot(iris_data['Id'], iris_data['petal.length'], label='Petal Length')
plt.xlabel('Id')
plt.ylabel('Petal Length')
plt.title('Line Plot: Petal Length vs Id')
plt.legend()
plt.show()

plt.bar(iris_data['Id'], iris_data['sepal.width'], label='Sepal Width')
plt.xlabel('Id')
plt.ylabel('Sepal Width')
plt.title('Bar Plot: Sepal Width vs Id')
plt.show()
```

## Visualizing Iris Variety Frequency Distribution

```python
# Bar Plot for Iris Varieties Frequency
frequency = iris_data['variety'].value_counts()
frequency.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Bar Plot: Frequency of Iris Varieties')
plt.ylabel('Count')
plt.show()

# Pie Chart for Iris Varieties Distribution (Fixed)

frequency = iris_data['variety'].value_counts()
colors = ['blue', 'green', 'red']
explode = (0.1, 0.1, 0.1)  # Slightly separate each section for clarity

frequency.plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    startangle=90,  # Rotate for better alignment
    textprops={'fontsize': 12}  # Adjust text size
)
plt.title('Pie Chart: Distribution of Iris Varieties', fontsize=14)
plt.ylabel('')  # Remove default y-axis label for better visuals
plt.show()
```

## Exploring Relationships Between Features


```python
# Scatter Plots for Sepal and Petal Dimensions
plt.scatter(iris_data['sepal.length'], iris_data['sepal.width'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()

plt.scatter(iris_data['petal.length'], iris_data['petal.width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.show()
```


## Advanced Scatter Plots with Seaborn

```python
# Seaborn Scatterplot: Sepal Dimensions by Variety
sb.scatterplot(data=iris_data, x='sepal.length', y='sepal.width', hue='variety')
plt.title('Seaborn Scatterplot: Sepal Dimensions by Variety')
plt.show()

# Seaborn Scatterplot: Petal Dimensions by Variety
sb.scatterplot(data=iris_data, x='petal.length', y='petal.width', hue='variety')
plt.title('Seaborn Scatterplot: Petal Dimensions by Variety')
plt.show()
```

## Heatmap Analysis of Feature Correlations

```python
# Overall Heatmap for Feature Correlations
frame_without_variety = iris_data.drop(columns=['variety', 'Id'])
sb.heatmap(frame_without_variety.corr(), annot=True)
plt.title('Heatmap: Feature Correlations (Overall)')
plt.show()

# Heatmaps by Iris Variety
grouped_data = iris_data.groupby('variety').corr()
for variety, group in grouped_data.groupby(level=0):
    sb.heatmap(group.droplevel(0), annot=True)
    plt.title(f'Heatmap: Feature Correlations ({variety})')
    plt.show()
```
